```python
from bs4 import BeautifulSoup as bs
import requests

import pandas as pd
```

```python
def format_product_id(product_id):
    product_id = product_id.lower()
    product_id = product_id.replace(" ","_")
    product_id = product_id.replace("'","")
    return product_id
```

```python
# jaspal
url = "https://www.jaspal.com/th/women/ready-to-wear/jaspal-x-kenny-scharf-blobz-r-me-t-shirt-1946084" #T shirt 
product_id = "Jaspal x Kenny Scharf Blobz'r Me T-shirt"
product_id = format_product_id(product_id)
# url = "https://www.jaspal.com/en/catalog/product/view/id/400366/s/floral-embroidered-mini-dress-1937538/category/928/"
# product_id = "Floral Embroidered Mini Dress"
# product_id = format_product_id(product_id)

# url = "https://www.homepro.co.th/p/1169585" #Samsung
# product_id = "refrigerator_samsung"
# url = "https://www.homepro.co.th/p/1214093" #Sharp
# product_id = "refrigerator_sharp"


DEBUG = False
DEBUG = True
product_id
```

```python

data_dict = {}
r = requests.get(url)
soup = bs(r.content, "html.parser")
if DEBUG:
    print(soup.prettify())
```

```python
# find h1 page title
page_title = soup.find('h1', class_='page-title').text
if DEBUG:
    print(page_title)
data_dict['product_name'] = page_title
# find span price
price = soup.find('span', class_='price').text
if DEBUG:
    print(price)
data_dict['price'] = price
```

```python
# find div class value itemprop = description
div_value = soup.find('div', class_='value', itemprop='description').text
if DEBUG:
    print(div_value)
data_dict['div_value'] = div_value
```

```python
# find div product-accordion
div_product_accordion = soup.find('div', class_='product-accordion')
if DEBUG:
    print(div_product_accordion)
```

```python
import traceback
# find div data-role="content" from div_product_accordion
try:
    
    material = div_product_accordion.find('div', attrs={'data-role':'content'}).text
    material = material.strip()
    if DEBUG:
        print(material)
    data_dict['material'] = material
except:
    print(traceback.format_exc())
    print("No material")
```

```python
import tiktoken
import os
def count_token(text, model="gpt-3.5-turbo"):
    encoding = tiktoken.encoding_for_model(model)

    return len(encoding.encode(text))
```

```python
sorted_keys = sorted(data_dict.keys(), key=lambda k:  count_token(data_dict[k])if isinstance(data_dict[k],str) else count_token("".join(data_dict[k])))

# Print the sorted keys
for key in sorted_keys:
    value = data_dict[key]
    try:
        print("key",key,count_token(value))
    except:
        print("key",key,count_token("".join(value)))
```

```python
# # Function to count tokens in Thai text
# def count_tokens(text):
#     tokens = text.split()  # Split the text into tokens (assuming space as a separator)
#     return len(tokens)

accumulated_token = 0
threshold_split_page = 500
page_data = ""
page_num = 0

for key in sorted_keys:
    value = data_dict[key]

    if isinstance(value, list):
        # If the value is a list, iterate over its items
        for item in value:
            token = count_token(item)
            accumulated_token += token

            # Check if accumulated tokens exceed the threshold
            if accumulated_token > threshold_split_page:
                accumulated_token = 0
                page_num += 1
                with open(f"{save_folder}/eng_{page_num}.txt", "w") as f:
                    f.write(page_data)
                page_data = ""

            page_data += f"{key}: {item}\n"
    elif isinstance(value, str):
        token = count_token(value)
        accumulated_token += token

        # Check if accumulated tokens exceed the threshold
        if accumulated_token > threshold_split_page:
            accumulated_token = 0
            page_num += 1
            with open(f"{save_folder}/eng_{page_num}.txt", "w") as f:
                f.write(page_data)
            page_data = ""

        page_data += f"{key}: {value}\n"
    else:
        raise Exception("Not supported type")

# Save any remaining data to the last page
if accumulated_token > 0:
    page_num += 1
    with open(f"{save_folder}/eng_{page_num}.txt", "w") as f:
        f.write(page_data)
```
